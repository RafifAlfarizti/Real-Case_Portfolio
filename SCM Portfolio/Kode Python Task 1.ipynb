# 5. Forecasting REVENUE 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import warnings
warnings.filterwarnings('ignore')
```

```python
# Mengkonversi kolom YEAR menjadi tipe datetime
data['YEAR'] = pd.to_datetime(data['YEAR'], format='%Y')
```

```python
# Mengagregasi data berdasarkan tahun
annual_revenue = data.groupby(data['YEAR'].dt.year)['REVENUE'].sum()
```

```python
# Membuat model ARIMA
model = ARIMA(annual_revenue, order=(1, 1, 1))
model_fit = model.fit()
```

```python
# Melakukan prediksi 5 tahun ke depan
forecast = model_fit.forecast(steps=5)
```

```python
# Menampilkan hasil prediksi
print(forecast)
```

Beginilah prediksi rata rata pendapatan pada 5 tahun yang akan datang
