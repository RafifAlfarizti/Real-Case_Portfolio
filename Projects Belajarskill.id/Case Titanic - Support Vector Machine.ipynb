# Data Preparation

```python
# Menghapus baris dengan nilai yang hilang
titanic = titanic.dropna()  # Drop baris dengan NaN di kolom penting
titanic
```

```python
# Menggunakan One-Hot Encoding untuk variabel kategorik
tc = titanic[["sex", "embarked", "class", "embark_town", "who"]]
tn = titanic[["age", "fare", "sibsp", "parch"]]
tn
```

# Splitting Data

```python
# Pisahkan fitur dan target
X = tn  # Fitur
y = titanic["survived"]  # Target
```

```python
# Pisahkan data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)

# Normalisasi data menggunakan StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(X_train_scaled)
```

# Fitting & Evaluation SVM Model

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Buat objek model SVM dengan kernel linear
model = SVC(kernel='linear')

# Lakukan fitting model dengan data latih
model.fit(X_train_scaled, y_train)

# Lakukan prediksi pada data uji
y_pred = model.predict(X_test_scaled)

# Evaluasi performa model dengan Confusion Matrix dan metrik evaluasi
cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Parameter Model:", model.get_params(), "\n")
print("Koefisien Model (Jika Linear):", model.coef_ if hasattr(model, "coef_") else "N/A", "\n")
print("Confusion Matrix:")
print(cm, "\n")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

model.get_params(): Menampilkan parameter default untuk SVM dengan kernel linear. Parameter ini bisa diatur untuk mengubah perilaku model.

model.coef_: Koefisien dari model SVM dengan kernel linear, koefisien ini memberikan informasi tentang orientasi dan bobot vektor pendukung. Dalam contoh ini,kita memiliki empat koefisien yang dapat membantu menginterpretasikan bagaimana fitur yang berbeda berkontribusi terhadap keputusan klasifikasi.

Confusion Matrix <br>
Baris pertama (kelas 0) menunjukkan bahwa semua 18 prediksi adalah kesalahan (false negatives), tidak ada true positives. <br>
Baris kedua (kelas 1) menunjukkan bahwa semua 37 prediksi adalah benar (true positives), tidak ada false negatives. <br>
Model ini sangat bias ke satu kelas (kelas 1), mungkin karena ketidakseimbangan data atau batasan dalam pemisahan linear.

Precision dan recall memberikan informasi tambahan yang berguna dalam kasus ketidakseimbangan kelas. Model ini memiliki recall tinggi, tetapi precision yang lebih rendah, menunjukkan banyak false positives. <br>
F1 score tinggi menunjukkan keseimbangan yang baik antara precision dan recall.

# Feature Selection & Regularization

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import GridSearchCV

# Buat objek model SVM
model = SVC()

# Gunakan SelectKBest untuk memilih 5 fitur terbaik
selector = SelectKBest(score_func=f_classif, k=4)
X_train_selected = selector.fit_transform(X_train_scaled, y_train)
X_test_selected = selector.transform(X_test_scaled)

# Lakukan Grid Search untuk mencari parameter C terbaik
param_grid = {'C': [0.1, 1, 10]}
grid_search = GridSearchCV(model, param_grid, cv=5)
grid_search.fit(X_train_selected, y_train)

# Hasil terbaik dari Grid Search
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Evaluasi performa model terbaik pada data uji
accuracy = best_model.score(X_test_selected, y_test)

print("Best Parameters:", best_params)
print("Accuracy on Test Data:", accuracy)
```

Hasil Grid Search menunjukkan bahwa regularisasi yang kuat (dengan C=0.1) adalah parameter terbaik untuk model SVM dengan 4 fitur terbaik.

Akurasi pada data uji menunjukkan bahwa model dapat bekerja dengan baik, namun masih ada ruang untuk perbaikan.

# Cross Validation

```python
from sklearn.model_selection import cross_val_score, GridSearchCV

model_cv = SVC(kernel='linear')
# Lakukan K-Fold Cross Validation dengan k=5
scores_cv = cross_val_score(model_cv, X_train_scaled, y_train, cv=5)

# Hitung rata-rata akurasi dari hasil Cross Validation
mean_accuracy_cv = scores_cv.mean()

print("Cross Validation Scores:", scores_cv)
print("Mean Accuracy from Cross Validation:", mean_accuracy_cv)
```

# Performance tunning

```python
model_tune = SVC()

# Tentukan parameter yang akan di-tune dan nilai-nilai yang akan dicoba
param_grid = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}

# Lakukan Grid Search untuk mencari parameter terbaik
grid_search = GridSearchCV(model_tune, param_grid, cv=5)
grid_search.fit(X_train_scaled, y_train)

# Hasil terbaik dari Grid Search
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Evaluasi performa model terbaik pada data uji
accuracy_tune = best_model.score(X_test_scaled, y_test)

print("Best Parameters:", best_params)
print("Best Model:", best_model)
print("Accuracy on Test Data (after tuning):", accuracy_tune)
```

Parameter terbaik untuk SVC (Support Vector Machine) dengan regularisasi yang lemah (C=10) dan kernel linear memberikan kinerja terbaik dalam Grid Search.
Dengan parameter tersebut, model memiliki akurasi sekitar 67.27% pada data uji, yang bisa menjadi indikator kinerja yang baik tetapi mungkin perlu dievaluasi lebih lanjut untuk menilai generalisasi.
